```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Ridge
```

```python
df2 = pd.read_csv('nbaoctgames2022current.csv', skiprows=1, names=['date', 'visitor', 'visitor_score', 'home', 'home_score']).drop(0)
df2.head()
```

```python
df2['score_difference'] = df2['home_score'] - df2['visitor_score']
```

```python
df2['home_win'] = np.where(df2['score_difference'] > 0, 1, 0)
df2['home_loss'] = np.where(df2['score_difference'] < 0, 1, 0)
```

```python
df2_visitor = pd.get_dummies(df2['visitor'], dtype=np.int64)
df2_home = pd.get_dummies(df2['home'], dtype=np.int64)
```

```python
# subtract home from visitor
df2_model = df2_home.sub(df2_visitor) 
df2_model['score_difference'] = df2['score_difference']
```

```python
df2_model['date'] = df2['date'].values
```

```python
df2_model.info()
```

```python
df2.head(5)
```

```python
df2_model['date'] = pd.to_datetime(df2_model['date'])
```

```python
# Chaning the date to run in LR = tunring into int from string
import datetime
df2_model['date']=df2_model['date'].map(datetime.datetime.toordinal)
```

```python
#Making new df with feature 'month' for LR model from string date to actual datetime value
df2_model['month'] = df2['date'].values
df2_model['month'] = pd.to_datetime(df2_model['month'])
df2_model['month'] = df2_model['month'].dt.month
```

```python
df2_model.info()
```

```python
df2_model.head()
```

```python
df_train = df2_model

lr = Ridge(alpha=0.001) 
X = df_train.drop(['score_difference'], axis=1)
y = df_train['score_difference']

lr.fit(X, y)
```

```python
df2_model_ratings = pd.DataFrame(data={'team': X.columns, 'rating': lr.coef_})
df2_model_ratings
```
